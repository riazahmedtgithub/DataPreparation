```python
import pandas as pd

# 1. Load the dataset as a Pandas data frame.
video_game = pd.read_csv(r'C:\Users\Riaz\Desktop\MSDS\Data Mining\Week 1\Video_Games_Sales_as_at_22_Dec_2016.csv')

# Required libraries are imported and using the read_csv function the dataframe has been loaded from csv file.
```

```python
# 2. Display the first ten rows of data.

# Displaying the first 10 rows of data by passing the argument as 10 to head function

print("The first 10 rows of video game data are,\n")
video_game.head(10)
```

```python
#3. Find the dimensions (number of rows and columns) in the data frame. 
#What do these two numbers represent in the context of the data?

print ("The rows of the dataframe are {}".format(video_game.shape[0]))
print ("The columns of the dataframe are {}".format(video_game.shape[1]))

# As seen from the below output, the first number indicates the number of rows and the second, the columns.
```

```python
# 4. Find the top five games by critic score.

# Using the sort_values function on the Critic_Score column and doing it descending wise,

video_game.sort_values(by='Critic_Score',ascending=False).head(5)
```

```python
# 5. Find the number of video games in the data frame in each genre.

print ("The number of video games in the data frame in each genre are\n")

video_game.groupby('Genre')['Name'].count()
```

```python
# 6. Find the first five games in the data frame on the SNES platform.


print("The first five games in the data frame on the SNES platform are\n")
video_game[video_game['Platform'] == 'SNES'].sort_values(by='Year_of_Release').head(5)

# First selecting the rows based on the value of Platform and then sorting it by using sort_values function to find 
# the releases in ascending order.
```

```python
# 7. Find the five publishers with the highest total global sales. 
# Note: You will need to calculate the total global sales for each publisher to do this.

print("Top five publishers with the highest total global sales.\n")

video_game.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(5)

# Groupby has been done on Publisher column and selecting Global_sales column to do a sum.  Then it has been sorted
```

```python
# 8. Create a new column in the data frame that calculates the percentage of global sales from North America. 
#Display the first five rows of the new data frame.

video_game['NA%Sales'] = (video_game['NA_Sales'] / video_game['Global_Sales'] ) * 100

print ("The first five rows of the new data frame with the new column NA%Sales, which is found at the end")

video_game.head(5)

```

```python
#9. Find the number NaN entries (missing data values) in each column.

print ('The number of NaN entries in each column.\n')

video_game.isna().sum()
```

```python
# 10. Try to calculate the median user score of all the video games. You will likely run into an error because some of 
# the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string 
# with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

# Trying to calculate the mean of User score column and getting the error

print (video_game['User_Score'].mean())

```

```python
# Getting the type of the data present in User_Score columns,

print ("The data type of User_Score column is\n ")

print (video_game['User_Score'].dtype,'\n')

# Getting the unique values present in that column.

print ('Unique values present in that column are \n')
print (video_game['User_Score'].unique(),'\n')

# Replacing the string with NaN and calculating the mean value

video_game['User_Score'] = pd.to_numeric(video_game['User_Score'],errors = 'coerce')

print ('Replacing the string with NaN and calculating the mean value\n')
print (pd.to_numeric(video_game['User_Score'],errors = 'coerce').mean(),'\n')

# Filling the NaN values with mean,

video_game['User_Score'].fillna(pd.to_numeric(video_game['User_Score'],errors = 'coerce').mean(),inplace=True)

# Confirming there are no NaN values after filling it with mean

print ("NaN values after filling it with mean for the column User_Score\n",video_game['User_Score'].isna().sum())

print ("The head output of final dataframe\n",video_game.head())
```
